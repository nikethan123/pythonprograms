```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mtcar_df = pd.read_csv("D:\AILAB\mtcars.csv")
mtcar_df
```

```python
temp = mtcar_df['am'].value_counts()
print(temp)
temp.plot.bar()
plt.title("Frequency distribution of transmssion types")
plt.xlabel("Transmission types")
plt.ylabel("Count")
```
